```python
import pandas as pd

# reformatting csvs

accuracies = pd.read_csv("/Users/cjz/Downloads/HC AI intern (Trevor)/ACTUAL/Task2/tests/Test 3_ Accuracies - Accuracy.csv")
accuracies.pop("Unnamed: 0")
accuracies = accuracies.dropna()
accuracies.rename(columns={"Unnamed: 32": "Accuracy"}, inplace=True)
accuracies = accuracies[['file_name', 'body', 'Accuracy', 'origin_location.location', 'origin_location.city', 'origin_location.state', 'origin_location.zip_code', 'origin_location.country', 'origin_location.country_code', 'origin_location.unlocode', 'destination_location.location', 'destination_location.city', 'destination_location.state', 'destination_location.zip_code', 'destination_location.country', 'destination_location.country_code', 'destination_location.unlocode', 'containers', 'commodity', 'incoterms', 'pickup_date', 'delivery_date', 'email_sender_information', 'hazardous', 'trailer', 'cargo_details.total_weights.weight', 'cargo_details.total_weights.unit', 'cargo_details.packages.type', 'cargo_details.packages.number_of_packages', 'cargo_details.packages.dimensions', 'cargo_details.total_volume.unit', 'cargo_details.total_volume.volume']]
# above changes order of columns so accuracy is closer to start

extractions = pd.read_csv("/Users/cjz/Downloads/HC AI intern (Trevor)/ACTUAL/Task2/tests/Test 3_ Extraction - Extraction.csv")
extractions["delete"] = False # just used for testing program, doesn't actually do anything

filename_pool = [accuracies.iloc[i, 0] for i in range(len(accuracies))] # pool of correct file names


dropped = 0
for i in range(len(extractions) - dropped): # loops through extractions csv to check which emails are included in accuracies csv; deletes all other emails from extractions csv
    if extractions.iloc[i - dropped, 0] not in filename_pool:
        extractions.iloc[i - dropped, -1] = True
        extractions = extractions.drop(i)
        dropped += 1
        
extractions.pop("delete")
```

```python
import csv

with open("/Users/cjz/Downloads/HC AI intern (Trevor)/ACTUAL/Task2/output.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(accuracies.columns)
    for row in range(len(accuracies)):
        writer.writerow(list(accuracies.iloc[row][0:3]) + list(extractions.iloc[row][2:]))
        
file.close()
```
